```python
#function to scroll from 0 position to end position
def scroll_to_bottom(driver):

    old_position = 0
    new_position = None

    while new_position != old_position:
        time.sleep(.5)

        # Get old scroll position
        old_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
        # Sleep and Scroll
        time.sleep(1)
        driver.execute_script((
                "var scrollingElement = (document.scrollingElement ||"
                " document.body);scrollingElement.scrollTop ="
                " scrollingElement.scrollHeight;"))
        # Get new position
        new_position = driver.execute_script(
                ("return (window.pageYOffset !== undefined) ?"
                 " window.pageYOffset : (document.documentElement ||"
                 " document.body.parentNode || document.body);"))
```

```python
#scraping html page data 
soup = BeautifulSoup(page_source, 'html.parser')
section=soup.findAll('div',{'class':'story_body_container'})
# print(section)
```

```python
section1 = soup.findAll('footer' ,{'class':'_22rc'})
# print(section1)
```

```python
for a in section:
    
    #for scraping post link and id
    link = a.find('a', attrs={'class':'_5msj'})
    post_link=link['href']
    part = post_link.split('&')[0]
    post_id=part.split('=')[1]
    post_links.append(post_link)
    post_ids.append(post_id)
    
    #for scraping date and time of post
    post_date=a.find('abbr')
    post_data=post_date.get_text(strip=True).split('at')
    date=post_data[0]
    time=post_data[-1]
    dates.append(date)
    times.append(time)
    
    #for scraping text of post
    text=a.find('div',{'class':'_5rgt _5nk5 _5msi'})
    post_text=text.find('div',{'class':'_2pim'})
    texts.append(post_text)
#     if(len(post_text)==-1):
#     post_text =" "
#     texts.append(post_text.get_text(strip=True))
#     description = a.text
#     print(description)
print(post_text)
print(date)
print(time)
```

```python
for a in section1:
    
    like=a.find('div', attrs={'class':'_1g06'})
    if(len(like)==0):
        like = "0 likes"
    likes.append(like.get_text(strip=True))
#     likes.append(like)
# print(like)
    #for scraping comment and share of post
    
    comm_shar=a.findAll('span',{'class':'_1j-c'})
    comments.append(comm_shar[0].get_text(strip=True))
    shares.append(comm_shar[1].get_text(strip=True))
print(comments)
print(shares)
    
```

```python
a = {'Posts_text': texts, 'Post_link_Id': post_links,'Shares':shares,'Date':dates,'Time':times,'Likes':likes,'Comments':comments}
df = pd.DataFrame.from_dict(a, orient='index')
df = df.transpose()
```
